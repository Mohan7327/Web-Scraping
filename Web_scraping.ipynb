```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/poco-m3-pro-5g-cool-blue-128-gb/product-reviews/itmf923739d27b4b?pid=MOBG3P4ZMCSJNSAY&lid=LSTMOBG3P4ZMCSJNSAYIYDCCL&marketplace=FLIPKART'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser') # depend on the page we should select the parser
print(soup.prettify()) #it remove the newline and display in the form of html
```

```python
scraped_revies=soup.find_all('p',class_='_2-N8zT')
scraped_revies
# note id (td) and class in website itself
```

```python
reviews=[]
for review in scraped_revies:
    review=review.get_text().replace('\n','') # replacing newline
    review=review.strip(' ') # to remove all white spaces, tabs and newline
    reviews.append(review)
reviews
```

```python
scraped_ratings=soup.find_all('div',class_='_3LWZlK _1BLPMq')
scraped_ratings
```

```python
ratings=[]
for rating in scraped_ratings:
    rating=rating.get_text().replace('\n','')
    rating=rating.strip(' ')
    ratings.append(rating)
ratings
```

```python
data=pd.DataFrame() # Dataframe
data['Comments'] = reviews
data['Ratings']=ratings
data
```

```python
data.to_csv('Flipkart Reviews.csv',index=False) # Exporting Data
```
